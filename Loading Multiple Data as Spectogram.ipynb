The first step is to load all of the required libraries

```python
# Import libraries for use on .wav files
import wave

# Import libraries for file management, locating and loading data.
import fnmatch
import os

# Pandas for data storage during processing
import pandas as pd

# for creating the spectogram
from scipy import signal
import matplotlib.pyplot as plt
import pylab

# location of the samples to be used
directory = './heartbeat-sounds/set_b/'
```

Next iterate over the files in the selected location an load the normal heartbeat data
displaying the name of each file as its loaded to check against the files in the location

```python
# Instatiate a dataframe to store multiple samples
normal_df = pd.DataFrame()

# iterate over all of the files in the specified folder and 
# load the normal data set
for filename in os.listdir(directory):
    if filename.endswith(".wav") and fnmatch.fnmatch(filename, '*normal__*'):
         # Get full filename by concatenating the path and filename, print out to indicate file will be loaded
        FNAME = os.path.join(directory, filename)
        print(FNAME)
        
        # Load file and unpack the samples
        f = wave.open(FNAME)
        frames = f.readframes(-1)
        sound_info = pylab.fromstring(frames, 'Int16')
        frame_rate = f.getframerate()

        # convert the samples to a spectrogram and then store in the dataframe
        spectrogram = pylab.specgram(sound_info, Fs=frame_rate)       
        temp_df = pd.DataFrame(list(spectrogram))
        normal_df = normal_df.append(temp_df.T,ignore_index=True)
        continue
    else:
        # ignore files that do not match specified criteria
        continue
        
```

spectrogram consists of; <br>
spectrum : 2-D array <br>
Columns are the periodograms of successive segments. <br>
freqs : 1-D array <br>
The frequencies corresponding to the rows in spectrum. <br>
t : 1-D array <br>
The times corresponding to midpoints of segments (i.e., the columns in spectrum).

Build the Neural Network model consisting of convolution and pooling layers to extract featured from the spectogram

```python
model = Sequential()
nb_classes = 10

model.add(Spectrogram(n_dft=512, n_hop=256, input_shape=input_shape, 
          return_decibel_spectrogram=True, power_spectrogram=2.0, 
          trainable_kernel=False, name='static_stft'))

model.add(Convolution2D(32,3,3, 
                        border_mode='same',
                        input_shape=(128,128,3)))
model.add(Activation('relu'))
model.add(Convolution2D(32,3,3))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Convolution2D(64,3,3, border_mode='same'))
model.add(Activation('relu'))
model.add(Convolution2D(64,3,3))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(nb_classes))
model.add(Activation('softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

Train and test the model

```python
batch_size = 128
nb_classes = 10
nb_epoch = 20

history = model.fit(df_normal_training, df_murmur_training,
                    batch_size=batch_size, nb_epoch=nb_epoch,
                    verbose=1, validation_data=(df_normal_test, df_murmur_test))
score = model.evaluate(df_normal_test, df_murmur_test, verbose=0)

print('Test score:', score)
print('Test accuracy:', score)
```
